```python
import tabula
import pandas as pd
import numpy as np
import math

# Inputs 

company_Name = "International Airline Group"
company_Ticker = "LON:IAG"

accounts_data_dict = {}
account_type_array = []
```

```python

df = tabula.read_pdf("IAG2019.pdf", pages=135)

print("")
print(df[0].head(5))
```

```python
print(df[0]["Unnamed: 0"])
```

```python
# Profit and Loss Finder

P_and_L_names = df[0]["Unnamed: 0"]
P_and_L_2019 = df[0]["Unnamed: 4"]
P_and_L_2018 = df[0]["Unnamed: 5"]

P_and_L_account_line_array = []
P_and_L_2019_value_array = []
P_and_L_2018_value_array = []

for counter, P_and_L_name in enumerate(P_and_L_names):

    try:
        d = math.isnan(P_and_L_name)
    except:
        
        if "total revenue" in P_and_L_name.lower():
            P_and_L_2019_value_array.append(P_and_L_2019[counter])
            P_and_L_2018_value_array.append(P_and_L_2018[counter])
            P_and_L_account_line_array.append(P_and_L_name)
        
        if "operating profit" in P_and_L_name.lower():
            P_and_L_2019_value_array.append(P_and_L_2019[counter])
            P_and_L_2018_value_array.append(P_and_L_2018[counter])
            P_and_L_account_line_array.append(P_and_L_name)
        
        if "profit before tax" in P_and_L_name.lower():
            P_and_L_2019_value_array.append(P_and_L_2019[counter])
            P_and_L_2018_value_array.append(P_and_L_2018[counter])
            P_and_L_account_line_array.append(P_and_L_name)
            
        if "finance costs" in P_and_L_name.lower():
            P_and_L_2019_value_array.append(P_and_L_2019[counter])
            P_and_L_2018_value_array.append(P_and_L_2018[counter])
            P_and_L_account_line_array.append(P_and_L_name)
            
        if "profit after tax" in P_and_L_name.lower():
            P_and_L_2019_value_array.append(P_and_L_2019[counter])
            P_and_L_2018_value_array.append(P_and_L_2018[counter])
            P_and_L_account_line_array.append(P_and_L_name)
            

accounts_data_dict["Account Line"] = P_and_L_account_line_array
accounts_data_dict["2019"] = P_and_L_2019_value_array
accounts_data_dict["2018"] = P_and_L_2018_value_array

print(accounts_data_dict)
```

```python
accounts_data_dict["Account Type"] = account_type_array
print(accounts_data_dict)
```

```python
cleaned_df = pd.DataFrame(data=accounts_data_dict)
print(cleaned_df)
```

```python

cleaned_df.to_csv(r'File.csv', index = False)
```
